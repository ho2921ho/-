# 2. 데이터 결합

```python
# 상반기와 하반기 데이터를 상하로 결합
df = pd.concat([first,second], axis = 0)
df
```

# 3. 데이터 가공 및 분석

```python
# 데이터 프레임 정보 조회
df.info()
```

```python
# 상표 컬럼의 고유값 조회
set(df['상표'])
```

```python
# 지역 컬럼의 고유값 조회
set(df['지역'])
```

```python
# 지역 컬럼의 값을 이용하여 시와 구를 분리
df['시'] = [x[0:2] for x in df['지역']]
df['구'] = [x[3:] for x in df['지역']]
df
```

```python
# 시 컬럼의 서울을 서울특별시로 변환
df['시'] = [x + '특별시' for x in df['시']]
df
```

```python
# 기간 컬럼을 datetime 형식으로 변경
df['기간'] = pd.to_datetime(df['기간'], format = '%Y%m%d')
df
```

```python
# 기간 컬럼에서 dt 타입을 이용하여 년, 월, 일, 요일 컬럼 생성
df['월'] = [x.month for x in df['기간']]
df['년'] = [x.year for x in df['기간']]
df['일'] = [x.day for x in df['기간']]
df['요일'] = [x.weekday() for x in df['기간']]
```

```python
weekday_dict = {0:'Monday',1:'Tuesday',2:'Wednesday', 3: 'Thursday',4:'Friday',5:'Saturday',6:'Sunday'}
```

```python
df['요일'] = [weekday_dict[x] for x in df['요일']]
```

```python
# 상표별 평균 휘발유 가격을 계산하세요.
mean_Hoil = df.groupby('상표')['휘발유'].mean().to_frame()
mean_Hoil
```

```python
# 상표별 평균 휘발유 가격을 시각화하세요.
ax = sns.barplot( y = '휘발유',x = mean_Hoil.index, data=mean_Hoil)
plt.title("상표별 평균 휘발유 가격")
plt.show()
```

```python
# 상표와 셀프여부별 평균 휘발유 가격을 계산하세요.
mean_Hoil_2 = df.groupby(['상표','셀프여부'])['휘발유'].mean().to_frame()
mean_Hoil_2
```

```python
# 휘발유 가격과 일반 주유소의 가격을 boxplot을 이용하여 시각화해보세요.
# 휘발유가 0인 데이터는 제외하고 시각화합니다.
plt.rcParams["figure.figsize"] = (12,6)
df2 = df[df['휘발유'] > 0]
ax = sns.boxplot(x="상표", y="휘발유", hue = '셀프여부', data=df2)
plt.show()
```

```python
# 지역별 평균 휘발유 가격을 계산하세요
df3 = df.groupby('구')['휘발유'].mean().to_frame().sort_values(by = '휘발유', ascending = False)
```

```python
# 지역별 평균 휘발유 가격을 시각화하세요.
ax = sns.barplot( y = '휘발유',x = df3.index, data=df3)
plt.title("지역별 평균 휘발유 가격")
plt.xticks(rotation=90)
plt.show()
```

```python
# 월별 평균 휘발유 가격을 계산하세요
df4 = df.groupby('월')['휘발유'].mean().to_frame()
df4
```

```python
# 월별 평균 휘발유 가격을 시각화 하세요.
ax = sns.barplot( y = '휘발유',x = df4.index, data=df4)
plt.title("월별 평균 휘발유 가격")
plt.show()
```
