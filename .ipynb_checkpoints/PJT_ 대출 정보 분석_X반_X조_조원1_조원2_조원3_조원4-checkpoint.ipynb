# 대출 정보 분석하기

```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
```

# 1. 고객 정보 파악하기

```python
# 고객 정보(data/고객정보.csv)를 customer에 불러오세요.

customer = pd.read_csv('./data/고객정보.csv', encoding = 'ms949')

customer
```

```python
# customer의 컬럼 정보를 확인해보세요.
customer.columns
```

```python
# customer의 컬럼 정보를 아래의 컬럼정보로 변경하세요.
#['회원번호', '직업', '추정소득', '가구추정소득', '실가족원수', '보험가입가족원수', '막내자녀나이','배우자직업','배우자추정소득','신용등급','나이','성별']
customer.columns = ['회원번호', '직업', '추정소득', '가구추정소득', '실가족원수', '보험가입가족원수', '막내자녀나이','배우자직업','배우자추정소득','신용등급','나이','성별']
```

```python
# 직업 Columns에 어떤 값들이 들어가있는지 확인해 보세요.
customer['직업'].unique()
```

```python
# 나이 Columns에 어떤 값들이 들어가있는지 확인해 보세요.
customer['나이'].unique()
```

```python
# 추정소득 Columns을 기준으로 내림차순 정렬 후 상위 10개만 출력해보세요.
customer.sort_values(by = '추정소득', ascending = False).head(10)
```

```python
# customer 데이터 프레임의 컬럼별 결측치(NA)의 개수를 출력하세요.
customer.isnull().sum()
```

```python
# 직업, 배우자 직업 Columns의 결측치(NA)를 *(정보없음)으로 치환하세요.
customer['직업'] = customer['직업'].fillna('*')
```

```python
# 막내자녀나이 Columns의 결측치(NA)를 0으로 치환하세요.
customer['막내자녀나이'] = customer['막내자녀나이'].fillna(0)
```

```python
# 나이 Columns의 *을 0으로 변경하세요.
customer['나이'] = customer['나이'].replace('*',0)
```

```python
# 나이 Columns의 데이터 타입을 int로 변경하세요
customer['나이'] = customer['나이'].astype(int)
```

# 2. 대출 정보 파악하기

```python
# 대출 정보(data/대출정보.csv)를 loan 불러오세요.
loan = pd.read_csv('./data/대출정보.csv', encoding = 'ms949')
loan
```

```python
# loan 컬럼 정보를 아래의 컬럼정보로 변경하세요.
#['회원번호', '연체여부', '현재 은행 대출 총 건수', '현재 카드사/할부사/캐피탈 대출 총 건수', '현재 2산업분류 대출 총 건수', '현재 기타 대출 총 건수', '현재 대출 총 금액','현재 신용대출 총 금액','현재 은행대출 총 금액','현재 카드사/할부사/캐피탈 총 금액','신용대출 유지기간', '2산업분류-신용대출 유지기간', '신용카드 개설 총 건수', '신용카드 유지기간', '보증정보 총 건수', '보증정보 총 금액']

loan.columns = ['회원번호', '연체여부', '현재 은행 대출 총 건수', '현재 카드사/할부사/캐피탈 대출 총 건수', '현재 2산업분류 대출 총 건수', '현재 기타 대출 총 건수', '현재 대출 총 금액','현재 신용대출 총 금액','현재 은행대출 총 금액','현재 카드사/할부사/캐피탈 총 금액','신용대출 유지기간', '2산업분류-신용대출 유지기간', '신용카드 개설 총 건수', '신용카드 유지기간', '보증정보 총 건수', '보증정보 총 금액']
loan
```

```python
# 연체 건수를 출력하세요.
loan['연체여부'].value_counts()
```

```python
# 현재 대출 총 금액 Columns을 기준으로 내림차순 정렬 후 상위 10개만 출력해보세요.
loan.sort_values(by = '현재 대출 총 금액', ascending = False).head(10)
```

```python
# 아래의 Columns 단위는 천원입니다. 단위를 만원으로 변환하세요.
# '현재 대출 총 금액', '현재 신용대출 총 금액', '현재 은행대출 총 금액', '현재 카드사/할부사/캐피탈 총 금액', '신용대출 유지기간', '2산업분류-신용대출 유지기간', '보증정보 총 금액'
columns = ['현재 대출 총 금액', '현재 신용대출 총 금액', '현재 은행대출 총 금액', '현재 카드사/할부사/캐피탈 총 금액', '신용대출 유지기간', '2산업분류-신용대출 유지기간', '보증정보 총 금액']
for col in columns:
    loan[col] = loan[col] / 1000
```

```python
loan
```

# 3. 두 데이터 프레임 병합 및 분석

```python
# customer데이터 프레임과 loan 데이터 프레임을 회원번호를 기준으로 병합 후 df에 저장하세요.
df = customer.merge(loan, how = 'left', on = '회원번호')
df
```

```python
# 직업을 기준으로 연체여부의 평균 계산 후 연체율이 높은 순서대로 출력하세요.
df.groupby('직업')['연체여부'].mean().to_frame().sort_values(by = '연체여부',ascending = False).rename({'연체여부':'mean'}, axis = 1)
```

```python
# 나이를 기준으로 연체여부의 평균 계산 후 연체율이 높은 순서대로 출력하세요.
df.groupby('나이')['연체여부'].mean().to_frame().sort_values(by = '연체여부',ascending = False).rename({'연체여부':'mean'}, axis = 1)
```

```python
# 성별을 기준으로 연체여부의 평균 계산 후 연체율이 높은 순서대로 출력하세요.
df.groupby('성별')['연체여부'].mean().to_frame().sort_values(by = '연체여부',ascending = False).rename({'연체여부':'mean'}, axis = 1)
```

```python
# 나이와 성별을 기준으로 연체여부의 평균 계산 후 연체율이 높은 순서대로 출력하세요.
df.groupby(['나이','성별'])['연체여부'].mean().to_frame().sort_values(by = '연체여부',ascending = False).rename({'연체여부':'mean'}, axis = 1)
```

```python
# 나이, 성별, 직업을 기준으로 연체여부의 평균 계산 후 연체율이 높은 순서대로 출력하세요.
df.groupby(['나이','성별','직업'])['연체여부'].mean().to_frame().sort_values(by = '연체여부',ascending = False).rename({'연체여부':'mean'}, axis = 1)
```

```python
# 직업별 평균 추정 소득과 평균 현재 대출 총 금액을 구해보세요.
df.groupby('직업')['추정소득','현재 대출 총 금액'].mean()
```

```python
# pivot_table 함수를 이용하여 직업과 성별을 기준으로 '현재 대출 총 금액', '현재 신용대출 총 금액', '현재 은행대출 총 금액', '현재 카드사/할부사/캐피탈 총 금액', '보증정보 총 금액'의 평균을 구해보세요. 
pd.pivot_table(df[['직업','성별','현재 대출 총 금액', '현재 신용대출 총 금액', '현재 은행대출 총 금액', '현재 카드사/할부사/캐피탈 총 금액', '보증정보 총 금액']], index = ['직업','성별'], aggfunc = 'mean')
```
